```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Dataset.csv')
df = pd.DataFrame(df)
df
```

```python
df0 = df[df.SepsisLabel == 0]
df0.dropna()
df0 = df0.fillna(df0.mean())
df0
```

```python
df1 = df[df.SepsisLabel == 1]
df1.dropna()
df1 = df1.fillna(df1.mean())
df1
```

```python
df1 = df1.sample(n=20000)
df0 = df0.sample(n=20000)
df2 = pd.concat([df1,df0])
df2.drop(['Hour','Patient_ID'], axis=1)
df2
```

```python
df2.to_csv('CleanDataset.csv')
```
