```python
import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier 
#to draw the graph
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus 
```

```python
import os
os.environ["PATH"] += os.pathsep + r'C:\Users\pavan\anaconda3\Library\bin\graphviz'
```

```python
data = pd.read_csv('CleanDataset.csv')
X = data.drop(["SepsisLabel"], axis = 1)
Y = data.SepsisLabel.values
```

```python
tree = DecisionTreeClassifier(criterion = 'entropy')
tree.fit(X,Y)
```

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=300, height=10)
```
