```python
import imblearn
import pandas as pd
from imblearn import under_sampling, over_sampling
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('Dataset.csv')
df = pd.DataFrame(df)
df
df.drop(['Hour','Patient_ID'], axis=1)
df = df.fillna(df.mean())
df
```

```python
X = df.drop(["SepsisLabel"], axis = 1)
y = df.SepsisLabel.values
```

```python
# transform the dataset
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
data = X
```

```python
data['SepsisLabel']=y
```

```python
data.to_csv('DatasetBalanced.csv')
```

```python
temp = data.Temp
hr =  data.HR
resp =  data.Resp
co2 = data.PaCO2
o2sat = data.O2Sat
y = data.SepsisLabel
tbl=data.loc[:,['Temp','HR','Resp','PaCO2','O2Sat','SepsisLabel']]
tbl
```

```python
true = 0
false = 0
output=[]
for i in range(0,len(y)):
    tally = 0    
    if(temp[i]>38 or temp[i]<36):
        tally +=1
    if(hr[i]>90):
        tally+=1
    if(resp[i]>20):
        tally+=1
    if(co2[i]>32):
        tally+=1
    if(o2sat[i]<95):
        tally+=1
    if(tally>=2 and y[i]==1) or (tally<2 and y[i]==0):
        true+=1
        output.append(1)
    else:
        false+=1
        #print(o2sat[i])
        output.append(0)
print("True count",true)
print("False count",false)
tbl['output']=output
tbl
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
confusion_matrix = metrics.confusion_matrix(tbl.SepsisLabel, tbl.output)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```
