# Module 2:- Understanding Sepsis and deriving rules

## Applying rules based approach

### Basic Rules form SIRS detection of sepsis any 2 conditions together are true indicates the person has sepsis
### 1. Temp > 38 or < 36
### 2. HR > 90
### 3. Resp >20
### 4. PaCO2 > 32mg

```python
#extracting the required columns
temp = data.Temp
hr =  data.HR
resp =  data.Resp
co2 = data.PaCO2
o2sat = data.O2Sat
y = data.SepsisLabel
tbl=data.loc[:,['Temp','HR','Resp','PaCO2','SepsisLabel']]
tbl
```

```python
output=[]
for i in range(0,len(y)):
    tally = 0    
    if(temp[i]>38 or temp[i]<36):
        tally +=1
    if(hr[i]>90):
        tally+=1
    if(resp[i]>20):
        tally+=1
    if(co2[i]>32):
        tally+=1
    if(tally>=2):
        output.append(1)
    else:
        output.append(0)

tbl['output']=output
tbl
```

```python
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(tbl.SepsisLabel, tbl.output)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
print(metrics.classification_report(tbl.SepsisLabel, tbl.output))
```

#### As per the confusion matrix is can be seen that there are many positives that are wrongly classified as well as there are majority of false instances classifed as ture
